## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to win the game.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible. In this way,all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, we have given length of word + 2 chances. For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.

```python
# Vocabulary of words
someWords = '''apple banana mango strawberry
orange grape pineapple apricot lemon coconut watermelon 
cherry papaya berry peach lychee muskmelon raspberry'''
```

```python
from random import choice

words = someWords.split()
word = choice(words)
```

```python
def print_list(word_list):
    for letter in word_list:
        print(letter, end=" ")
    print()
```

```python
def validate_input(guess, guessed_word):
    result = True
    if len(guess) != 1:
        print("Please enter a SINGLE LETTER")
        result = False
    elif not guess.isalpha():
        print("Please only enter ALPHABETS")
        result = False
    elif guess in guessed_word:
        print("You already guessed this letter. Try another one")
        result = False
    return result
```

```python
print("Welcome to Hangman! Guess a letter:")

guessed_word = ["_"] * len(word)

print_list(guessed_word)

# Define chances of wrong guess
chances = len(word) + 2

# Create flag for the game
playing = True

while playing:
    # Take input guess from the user
    guess = input("Enter a letter: ").lower()

    # Check the input is logically correct
    if not validate_input(guess, guessed_word):
        continue

    if guess in word:
        # Search for the index of the word
        for i in range(len(word)):
            if guess == word[i]:
                guessed_word[i] = guess

        # Check if the user won
        if "_" not in guessed_word:
            print_list(guessed_word)
            print("Congratulations! You won!")
            playing = False
            break
    else:
        # Reduce one chance
        chances -= 1

        # Check if the user lost
        if chances == 0:
            print("You lose! The word is: ", word)
            playing = False
            break
        print(f"Wrong guess. Try again! Chances left: {chances}")
        

    print_list(guessed_word)
```
